```python
#读取house或host的静态特征
def read_house_host(path, id_column, object, continuous_column, categorical_column1, categorical_column2,all_continuous_column=[]):
    # 从CSV文件中读取房屋信息
    data = pd.read_csv(path)
    # 创建重新编号的字典
    index = {id: new_id for new_id, id in enumerate(data[id_column].unique())}

    # 提取连续特征和类别特征,multi-hot特征
    continuous_features = data[continuous_column]
    categorical_features1 = data[categorical_column1]  # 类别特征列
    categorical_features2 = data[categorical_column2]

    # 使用LabelEncoder将类别特征转换为编号
    label_encoders = {}  # 存储特征列对应的编码器对象
    categorical_encodings = {}  # 存储特征值对应的编号
    for col in categorical_features2.columns:
        le = LabelEncoder()
        categorical_features2.loc[:, col] = le.fit_transform(categorical_features2[col])
        label_encoders[col] = le
        categorical_encodings[col] = dict(zip(le.transform(le.classes_), le.classes_))
    # 得到每个分类特征的类别数
    num_categories_list2 = [len(le.classes_) for le in label_encoders.values()]
    
    
    if object == 'house':  # house处理多热特征，host处理百分比连续特征
        multi_features = data.iloc[:, 10:52]
        #补缺失值，2
        multi_features=multi_features.fillna(-2)
        house_multi_matrix = np.array(multi_features)
        # 创建一个与最后一行维度相同的 -2 数组
        nan_row = np.full((1, house_multi_matrix.shape[1]),-2)
        # 将 NaN 数组插入到 house_multi_matrix 的最后一行
        house_multi_matrix = np.concatenate((house_multi_matrix, nan_row), axis=0)
        
        house_host_continuous_features = data[all_continuous_column]
        house_host_continuous_features=house_host_continuous_features.fillna(-2)
        house_host_continuous_matrix = np.array(house_host_continuous_features)
        nan_row = np.full((1, house_host_continuous_matrix.shape[1]),-2)
        house_host_continuous_matrix = np.concatenate((house_host_continuous_matrix, nan_row), axis=0)
    else:
        # 将连续特征中的百分比特征转换成数值类型
        # 找到包含百分数的列
        percent_columns = ['HOST_ACCEPTANCE_RATE', 'HOST_RESPONSE_RATE']
        # 将百分数转换为浮点数
        for column in percent_columns:
            continuous_features[column] = continuous_features[column].str.rstrip('%').astype(float) / 100.0

                        
    #补缺失值，连续值补-2，分类1特征补2，分类2特征不用补
    continuous_features=continuous_features.fillna(-2)
    categorical_features1=categorical_features1.fillna(2)
    
    # 将特征存放到矩阵中，每一列代表一个特征，特征值存放在对应的house编号索引中
    continuous_matrix = np.array(continuous_features)
    categorical_matrix1 = np.array(categorical_features1).astype(int)
    categorical_matrix2 = np.array(categorical_features2)

    # 添加一行-2
    nan_row = np.full((1, continuous_matrix.shape[1]),-2)
    continuous_matrix = np.vstack((continuous_matrix, nan_row))
    # 计算每一列的最大值加1
    new_row1 = np.full(categorical_matrix1.shape[1], 2)
    max_values2 = np.max(categorical_matrix2, axis=0) + 1
    # 将最大值加1的行添加到categorical_matrix1
    categorical_matrix1 = np.vstack((categorical_matrix1,new_row1))
    categorical_matrix2 = np.vstack((categorical_matrix2, max_values2))

    

    if object == 'house':  # house处理多热特征，host处理百分比连续特征
        return index, continuous_matrix,house_host_continuous_matrix, categorical_matrix1, categorical_matrix2, categorical_encodings, house_multi_matrix, num_categories_list2
    else:
        return index, continuous_matrix, categorical_matrix1, categorical_matrix2, categorical_encodings, num_categories_list2
```

```python
#读取host图片特征
def read_host_pic_features(path):
    data = pd.read_csv(path)
    # 创建重新编号的字典
    index = {id: new_id for new_id, id in enumerate(data['host_id'].unique())}
    # 提取连续特征
    continuous_features = data.iloc[:, 5:]
    host_pic_continuous_column = data.iloc[:, 5:].columns.tolist()
    #补缺失值-2
    continuous_features=continuous_features.fillna(-2)
        
    # 将特征存放到矩阵中，每一列代表一个特征，特征值存放在对应的house编号索引中
    continuous_matrix = np.array(continuous_features)

    #当host没有pic特征时，取最后一行，特征为-2
    # 创建全为nan的一行
    nan_row = np.full((1, continuous_matrix.shape[1]),-2)
    # 在continuous_matrix最后添加zero_row
    continuous_matrix = np.vstack((continuous_matrix,nan_row))

    return index,continuous_matrix,host_pic_continuous_column
```

```python
#读取house和host以及负样本动态特征
#读取house和host以及负样本动态特征
def read_comment_features(path,id_column,house_continuous_column,host_continuous_column):
    data = pd.read_csv(path)
    # 创建重新编号的字典
    index = {id: new_id for new_id, id in enumerate(data[id_column].unique())}
    # 提取连续特征
    house_continuous_features = data[house_continuous_column]
    host_continuous_features = data[host_continuous_column]
    #补缺失值-2
    house_continuous_features=house_continuous_features.fillna(-2)
    host_continuous_features=host_continuous_features.fillna(-2)
    
    # 将特征存放到矩阵中，每一列代表一个特征，特征值存放在对应的house编号索引中
    house_continuous_matrix = np.array(house_continuous_features)
    host_continuous_matrix = np.array(host_continuous_features)
    
    # 创建一个全为 -2 的行
    nan_row1 = np.full((1, house_continuous_matrix.shape[1]), -2)
    nan_row2 = np.full((1, host_continuous_matrix.shape[1]), -2)
    # 将 nan_row 插入到 host_continuous_matrix 的最后一行
    house_continuous_matrix = np.vstack((house_continuous_matrix, nan_row1))
    host_continuous_matrix = np.vstack((host_continuous_matrix, nan_row2))

    #提取对应的house_id和host_id
    comment_house=np.array(data['listing_id'])
    comment_host = np.array(data['host_id'])
    #最后一行为-1，comment_id不存在时，house_id,host_id为-1
    comment_house = np.append(comment_house, -1)
    comment_host = np.append(comment_host, -1)

    return index, house_continuous_matrix, host_continuous_matrix, comment_house,comment_host
```

```python
# 转换 history 列为长度为15的数组
def process_history(history,max_history_len):
    if len(history) >= max_history_len:
        processed_history = history[-max_history_len:]
    else:
        processed_history = [-1] * (max_history_len - len(history)) + history
    return processed_history


# 将填充-1的位置标记为True
def create_mask(history):
    mask = [True if item == -1 else False for item in history]
    return mask

#获取所有用户历史记录对应的house、host索引
def user_history(path, comment_house, comment_host, comment_index, house_index, host_index, host_pic_index,max_history_len):
    data = pd.read_csv(path)
    # 将history列的数据转换为列表格式
    data['history'] = data['history'].apply(lambda x: [int(item) for item in x.strip('[]').split()])
    # 获取唯一的 reviewer_id
    unique_reviewer_ids = data['reviewer_id'].unique()

    # 创建 reviewer_id 到编号的映射字典
    user_index = {reviewer_id: idx for idx, reviewer_id in enumerate(unique_reviewer_ids)}
    # 应用处理函数到每个 history
    data['processed_history'] = data['history'].apply(lambda x: process_history(x, max_history_len))  # shape（user_num,history_num）,comment_id,填充为-1
    # 生成整个 history 的矩阵
    history_matrix = np.array(data['processed_history'].tolist())

    # 将 history_matrix 中的 comment_id 转换为 host_id，不存在的设为 -1
    history_matrix_host = []
    history_matrix_house = []
    history_matrix_host_pic = []
    history_matrix_mask = []
    history_matrix_comment = []
    for user_history in history_matrix:
        user_history_host = []
        user_history_host_pic = []
        user_history_house = []
        user_history_mask = []
        user_history_comment = []

        for comment_id in user_history:
            user_history_comment.append(comment_index.get(comment_id, -1))  # comment编号，填充的comment编号为-1
            user_history_host_id = comment_host[comment_index.get(comment_id, -1)]  # host_id
            user_history_house_id = comment_house[comment_index.get(comment_id, -1)]  # house_id
            user_history_host.append(host_index.get(user_history_host_id, -1))  # host表编号
            user_history_host_pic.append(host_pic_index.get(user_history_host_id, -1))  # host_pic表编号
            user_history_house.append(house_index.get(user_history_house_id, -1))  # house表编号

        history_matrix_comment.append(user_history_comment)  # shape（user_num,history_num）,comment编号
        history_matrix_mask.append(create_mask(user_history))# 在填充的位置标记true
        history_matrix_host.append(user_history_host)  # host表编号,shape（user_num,history_num）
        history_matrix_host_pic.append(user_history_host_pic)  # host_pic表编号
        history_matrix_house.append(user_history_house)  # house表编号  shape（user_num,history_num）

    return user_index, history_matrix_comment, history_matrix_host, history_matrix_host_pic, history_matrix_house, history_matrix_mask
```

```python
#获得测试集正负样本对应的house、host索引,动态特征
def get_train_input_index(path,comment_path,npratio,house_comment_scaler,host_comment_scaler,comment_index, house_index, host_index, host_pic_index,user_index,comment_house,comment_host,house_comment_tensor,host_comment_tensor):
    neg_data=pd.read_csv(path)   #负样本特征表，train_neg_comment_path
    df_comment=pd.read_csv(comment_path)  #评论表
    #读取负样本动态特征，房子房东id，以及对应的正样本的评论id
    pos_comment_id=[]
    neg_house_comment_features=[]
    neg_host_comment_features=[]
    neg_house_id=[]
    neg_host_id=[]
    grouped_neg_data = neg_data.groupby('pos_comment_id')
    for name, group in grouped_neg_data:
        pos_comment_id.append(name)  #正样本comment_id
        neg_house_comment_matrix=group[house_comment_continuous_column].values[:]
        neg_host_comment_matrix=group[host_comment_continuous_column].values[:]
        
        #先填缺失值，再标准化
        neg_house_comment_matrix = np.nan_to_num(neg_house_comment_matrix, nan=-2)
        neg_host_comment_matrix = np.nan_to_num(neg_host_comment_matrix, nan=-2)
        
        #标准化
        neg_house_comment_matrix = house_comment_scaler.transform(neg_house_comment_matrix)
        neg_host_comment_matrix = host_comment_scaler.transform(neg_host_comment_matrix)
        
  
        neg_house_comment_features.append(neg_house_comment_matrix)  #负样本房子评论特征(6,特征数)
        neg_host_comment_features.append(neg_host_comment_matrix)#负样本房东评论特征
        neg_house_id.append(group['listing_id'].values[:])  #负样本房子id
        neg_host_id.append(group['host_id'].values[:])
        

    
    #负样本特征转换成tensor
    neg_house_comment_features=torch.tensor(neg_house_comment_features).clone().detach()
    neg_host_comment_features=torch.tensor(neg_host_comment_features).clone().detach()
    
    #根据id，获得负样本房子，房东的index
    neg_house_id=np.array(neg_house_id)
    neg_host_id=np.array(neg_host_id)
    # 使用NumPy的vectorized操作获取所有id对应的索引
    input_neg_house_index= np.vectorize(house_index.get)(neg_house_id)
    input_neg_host_pic_index= np.vectorize(host_pic_index.get,otypes=[int])(neg_host_id,-1)
    input_neg_host_index = np.vectorize(host_index.get, otypes=[int])(neg_host_id, -1)
    
    #根据正样本的评论id，获得房子，房东。comment。reviewer的index
    #获取正样本的评论id对应的index
    input_pos_comment_index=np.vectorize(comment_index.get)(pos_comment_id)
    #获得房子，房东。reviewer的id
    input_pos_house_id=comment_house[input_pos_comment_index]
    input_pos_host_id=comment_host[input_pos_comment_index]
    input_reviewer_id=df_comment['reviewer_id'][input_pos_comment_index].to_numpy()
    #获得房子，房东。reviewer的index
    input_pos_house_index=np.vectorize(house_index.get)(input_pos_house_id)
    input_pos_host_index=np.vectorize(host_index.get)(input_pos_host_id)
    input_pos_host_pic_index=np.vectorize(host_pic_index.get,otypes=[int])(input_pos_host_id,-1)
    input_reviewer_index=np.vectorize(user_index.get)(input_reviewer_id)


    #获得正样本的动态特征
    pos_house_comment_features =house_comment_tensor[input_pos_comment_index]
    pos_host_comment_features =host_comment_tensor[input_pos_comment_index]
    
    #合并正负样本
    #动态特征
    house_comment_features = torch.cat((pos_house_comment_features.unsqueeze(1), neg_house_comment_features), dim=1)
    host_comment_features = torch.cat((pos_host_comment_features.unsqueeze(1), neg_host_comment_features), dim=1)
    #房子，房东的index
    input_house_index=np.concatenate((input_pos_house_index[:, np.newaxis], input_neg_house_index), axis=1)
    input_host_index=np.concatenate((input_pos_host_index[:, np.newaxis], input_neg_host_index), axis=1)
    input_host_pic_index=np.concatenate((input_pos_host_pic_index[:, np.newaxis], input_neg_host_pic_index), axis=1)
    
    label = np.zeros((len(input_house_index), 1 + npratio))
    label[:, 0] = 1
    label= torch.tensor(label)
    
    return input_house_index,input_host_index,input_host_pic_index,input_reviewer_index,label,house_comment_features,host_comment_features
```

```python
#根据house索引，获得house的特征，返回为tensor
def get_house_input(house_continuous_tensor,house_host_continuous_tensor,house_categorical_tensor1,house_categorical_tensor2,house_multi_tensor,input_house_index):
    # 样本编号转换成tensor
    input_house_index = torch.tensor(input_house_index).clone().detach()
    # house
    # continuous_features
    house_continuous_features = house_continuous_tensor[input_house_index]
    house_host_continuous_features = house_host_continuous_tensor[input_house_index]
    # categorical_features1
    house_categorical_features1 = house_categorical_tensor1[input_house_index]
    # categorical_features2
    house_categorical_feature2 = house_categorical_tensor2[input_house_index]
    # multi_features
    house_multi_features = house_multi_tensor[input_house_index]

    return house_continuous_features,house_host_continuous_features,house_categorical_features1,house_categorical_feature2,house_multi_features
```

```python
def get_host_input(host_continuous_tensor,host_categorical_tensor1,host_categorical_tensor2,host_pic_continuous_tensor,input_host_index, input_host_pic_index):
    # host
    # continuous_features
    host_continuous_features = host_continuous_tensor[input_host_index]
    # categorical_features1
    host_categorical_features1 = host_categorical_tensor1[input_host_index]
    # categorical_features2
    host_categorical_features2 = host_categorical_tensor2[input_host_index]
    # host_pic_features
    host_pic_features = host_pic_continuous_tensor[input_host_pic_index]
            
    return host_continuous_features,host_categorical_features1,host_categorical_features2,host_pic_features
```

```python
#根据样本中的user编号获得对应house、host编号
def get_user_input_index(history_matrix_comment,history_matrix_host,history_matrix_host_pic,history_matrix_house,history_matrix_mask,input_reviewer_index):
    input_history_comment = history_matrix_comment[input_reviewer_index]  # shape(comment_num, max_history)
    input_history_host = history_matrix_host[input_reviewer_index]
    input_history_host_pic = history_matrix_host_pic[input_reviewer_index]
    input_history_house = history_matrix_house[input_reviewer_index]
    input_history_mask = history_matrix_mask[input_reviewer_index]

    return input_history_comment,input_history_host,input_history_host_pic,input_history_house,input_history_mask
```

```python
def creat_scaler(path,column):
    data=pd.read_csv(path)
    continuous_features=data[column]
    # 补缺失值-2
    continuous_features= continuous_features.fillna(-2)
    continuous_matrix=np.array(continuous_features)
    # 创建一个StandardScaler对象
    scaler = StandardScaler()
    scaler.fit(continuous_matrix)
    return scaler
```

```python
def create_centered_distance(X):
    # 计算每个样本的平方模长
    r = torch.sum(X ** 2, dim=1, keepdim=True)

    # 计算欧氏距离矩阵
    D = torch.sqrt(torch.clamp(r - 2 * torch.mm(X, X.t()) + r.t(), min=0.0) + 1e-8)

    # 计算 D 的中心化版本
    mean_axis_0 = torch.mean(D, dim=0, keepdim=True)
    mean_axis_1 = torch.mean(D, dim=1, keepdim=True)
    D1 = D - mean_axis_0 - mean_axis_1 + torch.mean(D)
    return D1
def create_distance_covariance(D1, D2):
    # 计算 D1 和 D2 之间的距离协方差
    n_samples = float(D1.size(0))
    dcov = torch.sqrt(torch.clamp(torch.sum(D1 * D2) / (n_samples * n_samples), min=0.0) + 1e-8)
    return dcov
def create_distance_correlation( X1, X2):
    X1=X1.reshape(-1,feature_dim)
    X2=X2.reshape(-1,feature_dim)
    D1 = create_centered_distance(X1)
    D2 = create_centered_distance(X2)

    dcov_12 = create_distance_covariance(D1, D2)
    dcov_11 = create_distance_covariance(D1, D1)
    dcov_22 = create_distance_covariance(D2, D2)

    dcor = dcov_12 / (torch.sqrt(torch.clamp(dcov_11 * dcov_22, min=0.0)) + 1e-10)
    return dcor
```

```python
import math

import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn import functional as F

#多头自注意力
class MultiHeadSelfAttention(nn.Module):
    def __init__(self, num_heads, feature_dim, max_history_len):
        super(MultiHeadSelfAttention, self).__init__()
        self.num_heads = num_heads  #10
        self.feature_dim = feature_dim   #200
        self.head_dim = feature_dim // num_heads
        self.max_history_len = max_history_len

        self.WQ = nn.Linear(feature_dim, feature_dim)
        self.WK = nn.Linear(feature_dim, feature_dim)
        self.WV = nn.Linear(feature_dim, feature_dim)

    def forward(self, history_matrix, mask=None):
        batch_size, history_len, _ = history_matrix.size()

        Q = self.WQ(history_matrix)
        K = self.WK(history_matrix)
        V = self.WV(history_matrix)

        Q = Q.view(batch_size, history_len, self.num_heads, self.head_dim).permute(0, 2, 1, 3)  #(batch,num_heads,history_len,head_dim)
        K = K.view(batch_size, history_len, self.num_heads, self.head_dim).permute(0, 2, 1, 3)
        V = V.view(batch_size, history_len, self.num_heads, self.head_dim).permute(0, 2, 1, 3)

        attention_scores = torch.matmul(Q, K.permute(0, 1, 3, 2)) / (self.head_dim ** 0.5)

        if mask is not None:
            attention_scores = attention_scores.masked_fill(mask.unsqueeze(1).unsqueeze(2), float('-inf'))  #()

        attention_weights = torch.softmax(attention_scores, dim=-1)  #shape(batch,head,history_len,history_len)
        #(batch,history_len,200)
        weighted_sum = torch.matmul(attention_weights, V).permute(0, 2, 1, 3).contiguous().view(batch_size, history_len,
                                                                                                self.feature_dim)

        return weighted_sum

#简化的加权注意力机制
class AttentivePooling(nn.Module):
    def __init__(self, dim1, dim2):
        super(AttentivePooling, self).__init__()
        self.dim1 = dim1   #history_len
        self.dim2 = dim2   #feature_dim=200
        self.dropout = nn.Dropout(0.2)
        self.dense1 = nn.Linear(dim2, 100)
        self.dense2 = nn.Linear(100, 1)

    def forward(self, input_matrix, mask=None):
        user_vecs = self.dropout(input_matrix) # （batch，15,200）
        user_att = self.dense2(F.relu(self.dense1(user_vecs))) # （batch，15，1）
        #得到权重
        user_att = user_att.view(-1, self.dim1)  # 展平为一维向量（batch，15）

        if mask is not None:
            user_att = user_att.masked_fill(mask, float('-inf'))  # 对填充向量进行mask操作

        user_att = F.softmax(user_att, dim=-1)  # 归一化操作

        user_vec = torch.bmm(user_att.unsqueeze(1), user_vecs).squeeze(1)  # 加权池化操作，（样本数，200）
        return user_vec

#用户建模
class MultiHeadAttentionAndPooling(nn.Module):
    def __init__(self, num_heads, feature_dim, max_history_len):
        super(MultiHeadAttentionAndPooling, self).__init__()
        self.multi_head_attention = MultiHeadSelfAttention(num_heads, feature_dim, max_history_len)  #(batch,history_len,feature_dim)
        self.attentive_pooling = AttentivePooling(max_history_len, feature_dim)    #(batch,feature_dim)

    def forward(self, history_matrix, mask=None):
        weighted_sum = self.multi_head_attention(history_matrix, mask)
        pooled_vector = self.attentive_pooling(weighted_sum, mask)
        return pooled_vector



def categorical1_embedding(num_categorical1_features): #输入categorical1的特征数
    # 创建一个列表来存储每个嵌入层
    categorical1_embeddings = []
    for i in range(num_categorical1_features):
        embedding_layer1 = nn.Embedding(3, 2)
        categorical1_embeddings.append(embedding_layer1)
    return categorical1_embeddings

def categorical2_embedding(num_categorical2_features,num_categories_list2): #输入categorical2的特征数，各个特征的特征值数量的列表
    # 创建一个列表来存储每个嵌入层
    categorical2_embeddings = []
    # 循环遍历每个特征，为每个特征创建嵌入层并添加到列表中
    for i in range(num_categorical2_features):
        categorie2_embedding_dim = int(math.log2(num_categories_list2[i]+1))  # 嵌入维度为类别数的底数为2的对数
        embedding_layer2 = nn.Embedding(num_categories_list2[i]+1, categorie2_embedding_dim)
        categorical2_embeddings.append(embedding_layer2)
    return categorical2_embeddings

 # 全连接层
def dense_layer(in_features, out_features):
    # in_features=hidden_size,out_features=1
    return nn.Sequential(
        nn.Linear(in_features, out_features,bias=True),
        nn.ReLU())

#dnn层,hidden_units=[[输入1，输出1],[输入2，输出2]]
class DNN(nn.Module):
    def __init__(self, hidden_units, activation, l2_reg, dropout_rate, use_bn):
        super(DNN, self).__init__()

        # 创建一个空的列表用于存储神经网络的层
        layers = []

        # 遍历每个隐藏层的设置
        for hidden_size in hidden_units:
            # 添加一个线性层（全连接层），输入维度为 hidden_size[0]，输出维度为 hidden_size[1]
            layers.append(nn.Linear(hidden_size[0], hidden_size[1]))

            # 如果 use_bn 为 True，则添加批归一化层
            if use_bn:
                layers.append(nn.BatchNorm1d(hidden_size[1]))

            # 添加激活函数层，激活函数为传入的 activation 函数
            layers.append(activation)

            # 添加 dropout 层，dropout_rate 为 dropout 的比例
            layers.append(nn.Dropout(dropout_rate))

        # 将所有层组合成一个顺序的神经网络
        self.dnn_network = nn.Sequential(*layers)

    def forward(self, x):
        # 在前向传播过程中，输入数据 x 经过 dnn_network 这个神经网络模型
        return self.dnn_network(x)

#多任务网络
class MMoE(nn.Module):
    def __init__(self, num_experts, num_tasks, expert_dnn_hidden_units, gate_dnn_hidden_units):
        super(MMoE, self).__init__()

        # 创建 num_experts 个专家网络，每个专家网络都是一个 DNN 模型
        self.expert_networks = nn.ModuleList([
            DNN(expert_dnn_hidden_units, nn.ReLU(), 0.0, 0.0, False) for _ in range(num_experts)
        ])

        # 创建 num_tasks 个门控网络，每个门控网络也是一个 DNN 模型
        self.gate_networks = nn.ModuleList([
            DNN(gate_dnn_hidden_units, nn.ReLU(), 0.0, 0.0, False) for _ in range(num_tasks)
        ])

        # 定义一个 expert_concat 变量，目前它是一个 torch.stack 函数，但是没有被调用
        self.expert_concat = torch.stack

    def forward(self, x): #x->house表示
        # 对于每个专家网络，将输入 x 传递给专家网络，得到专家的输出
        expert_outputs = [expert_network(x) for expert_network in self.expert_networks]   #len(list)=3,[tensor(样本数,15,200),（样本数,15,200),（样本数,15,200)]

        # 将专家的输出拼接在一起，形成一个张量
        expert_concat = torch.stack(expert_outputs, dim=2) #[样本数, 15, 3, 200]

        # 定义一个空列表，用于存储最终的 MMoE 输出
        mmoe_outputs = []

        # 遍历每个门控网络
        for i in range(len(self.gate_networks)):
            # 将输入 x 传递给门控网络，得到门控的输出
            gate_input = self.gate_networks[i](x)  #[样本数, 15, 3]

            # 使用 softmax 函数对门控的输出进行归一化，得到门控的权重
            gate_out = F.softmax(gate_input, dim=-1).unsqueeze(-1)  #[样本数, 15, 3, 1]

            # 将专家的输出与门控的权重相乘，得到加权后的专家输出
            gate_mul_expert = torch.sum(expert_concat * gate_out, dim=2)   #[样本数, 15, 200]

            # 将加权后的专家输出添加到 mmoe_outputs 列表中
            mmoe_outputs.append(gate_mul_expert)  #list[tensor(样本数, 15, 200),tensor(样本数, 15, 200)]

        # 返回最终的 MMoE 输出列表
        return mmoe_outputs


class Gate(nn.Module):
    def __init__(self,weight_dnn_hidden_units):
        super(Gate, self).__init__()

        self.gate = DNN(weight_dnn_hidden_units,nn.Tanh(), 0.0, 0.0, True)
        self.temp_gate=nn.Linear(2, 1)

    def forward(self, user,house):

        # 将 user_input 扩展为 [样本数, 6, 200]
        user=user.unsqueeze(1)
        user = user.expand(-1,npratio+1 , -1)
        # 合并 user_input 和 house_input 在最后一维上
        combined_vec = torch.cat((user, house), dim=-1)  #[样本数, 6, 400]
        combined_vec_flattened = combined_vec.view(-1, 600) #[样本数*6, 600]
        input_norm = F.normalize(combined_vec_flattened, p=2, dim=-1)
        gate_input = self.gate(input_norm)  #[样本数*6, 2]
        gate_temp=self.temp_gate(gate_input)#[样本数*6, 1]
        weight1 = torch.sigmoid(gate_temp).view(-1,npratio+1,1) #[样本数,6, 1）
#         weight1=self.gate(combined_vec)  #[样本数, 2, 1]
        ones_tensor = torch.ones_like(weight1)
        weight2=ones_tensor - weight1
        gate_output=torch.cat((weight1, weight2), dim=-1)#[样本数, 6, 2]
        return gate_output


class HouseEmbedding(nn.Module):
    def __init__(self,num_house_categorical1_features,num_house_categorical2_features,num_house_categories_list2,num_house_multi_features,multi_embedding_dim,feature_dim):
        super(HouseEmbedding, self).__init__()
        self.multi_embedding_dim=multi_embedding_dim
        self.num_house_multi_features=num_house_multi_features
        # house_categorical1
        self.house_categorical1_embeddings = categorical1_embedding(num_house_categorical1_features)
        #house_categorical2
        self.house_categorical2_embeddings = categorical2_embedding(num_house_categorical2_features,num_house_categories_list2)
        #house_multi
        self.multi_embedding_layer = dense_layer(num_house_multi_features, multi_embedding_dim)
        # MLP
        self.house_dense_layer = dense_layer(35, feature_dim)

    def forward(self, house_continuous_features,house_categorical_features1,house_categorical_feature2,house_multi_features):
        # house
        # house_continuous_features(样本数, 2, 14)
        # house_multi_features(样本数, 2, 42->18),(样本数, 15, 42->18)
        # 对多值特征进行嵌入
        temp_multi_embeddings = self.multi_embedding_layer(
            house_multi_features.view(-1, self.num_house_multi_features).float())  # shape(样本数*（负样本+1），18)
        # 将嵌入结果重塑为（样本数，正负样本数，嵌入维度）的形状
        house_multi_embeddings = temp_multi_embeddings.view(house_multi_features.shape[0],
                                                            house_multi_features.shape[1], self.multi_embedding_dim)
        # house_comment_features（样本数, 2, 18）
        # house_categories_embeddings1,shape(样本数，2,2(2*1))
        house_categories_embeddings1 = torch.cat(
            [embedding_layer(house_categorical_features1[:,:, i]) for i, embedding_layer in
             enumerate(self.house_categorical1_embeddings)], dim=-1)
        # house_categories_embeddings2，shape(样本数，2,(52+4)->embedding维度之和=7(5+2))
        house_categories_embeddings2 = torch.cat(
            [embedding_layer(house_categorical_feature2[:, :, i]) for i, embedding_layer in
             enumerate(self.house_categorical2_embeddings)], dim=-1)
        # concate（样本数, 2,61）
        house_con = torch.cat(
            [house_continuous_features, house_multi_embeddings, house_categories_embeddings1,
             house_categories_embeddings2], dim=-1)
        house_vec = self.house_dense_layer(house_con.float())  # shape(样本数, 2,200)
        return house_vec
    
class HouseHostEmbedding(nn.Module):
    def __init__(self,feature_dim):
        super(HouseHostEmbedding, self).__init__()

        # MLP
        self.house_dense_layer = dense_layer(26, feature_dim)

    def forward(self, house_host_continuous_features,house_comment_features):
        # house
        # house_continuous_features(样本数, 2, 14)
        # house_comment_features（样本数, 2, 18）
        # concate（样本数, 2,61）
        house_host_con = torch.cat(
            [house_host_continuous_features, house_comment_features], dim=-1)
        house_host_vec = self.house_dense_layer(house_host_con.float())  # shape(样本数, 2,200)
        return house_host_vec
    

class HostEmbedding(nn.Module):
    def __init__(self,num_host_categorical1_features,num_host_categorical2_features,num_host_categories_list2,feature_dim):
        super(HostEmbedding, self).__init__()
        # host_categorical1
        self.host_categorical1_embeddings = categorical1_embedding(num_host_categorical1_features)
        # host_categorical2
        self.host_categorical2_embeddings = categorical2_embedding(num_host_categorical2_features,num_host_categories_list2)
        # MLP
        self.host_dense_layer = dense_layer(195, feature_dim)

    def forward(self,host_continuous_features, host_categorical_features1, host_categorical_features2,host_comment_features, host_pic_features):
        # host_categories_embeddings1，shape(样本数，2,10(2*5))
        host_categories_embeddings1 = torch.cat(
            [embedding_layer(host_categorical_features1[:, :, i]) for i, embedding_layer in
             enumerate(self.host_categorical1_embeddings)], dim=-1)
        # host_categories_embeddings2，shape(样本数，2,5->2(2*1))
        host_categories_embeddings2 = torch.cat(
            [embedding_layer(host_categorical_features2[:, :, i]) for i, embedding_layer in
             enumerate(self.host_categorical2_embeddings)], dim=-1)
        # concate(样本数，2,193)
        host_con = torch.cat(
            [host_continuous_features,host_comment_features,host_pic_features, host_categories_embeddings1,
             host_categories_embeddings2], dim=-1)
        host_vec = self.host_dense_layer(host_con.float())
        return host_vec
```

```python
class MatchingModel(nn.Module):
    def __init__(self, num_house_categorical1_features,num_host_categorical1_features,num_house_categorical2_features,num_host_categorical2_features,num_house_multi_features,multi_embedding_dim,
                 num_house_categories_list2,num_host_categories_list2,feature_dim,num_heads,max_history_len,
                 num_experts, num_tasks, expert_hidden_units, gate_hidden_units,mmoe_dnn_hidden_units,user_dnn_hidden_units,house_dnn_hidden_units,weight_dnn_hidden_units):
        super(MatchingModel, self).__init__()
        #embedding
        self.house_embedding=HouseEmbedding(num_house_categorical1_features,num_house_categorical2_features,num_house_categories_list2,num_house_multi_features,multi_embedding_dim,feature_dim)
        self.host_embedding=HostEmbedding(num_host_categorical1_features,num_host_categorical2_features,num_host_categories_list2,feature_dim)
        self.house_host_embedding=HouseHostEmbedding(feature_dim)
        #MMOE
        self.mmoe_model = MMoE(num_experts, num_tasks, expert_hidden_units, gate_hidden_units)
        self.mmoe_mlp1=DNN(mmoe_dnn_hidden_units, nn.ReLU(), 0.0, 0.0, False)
        self.mmoe_mlp2=DNN(mmoe_dnn_hidden_units, nn.ReLU(), 0.0, 0.0, False)
        #attetion
        self.user_house_encoder = MultiHeadAttentionAndPooling(num_heads, feature_dim, max_history_len)
        self.user_host_encoder = MultiHeadAttentionAndPooling(num_heads, feature_dim, max_history_len)
        
        self.house_concate_dnn = DNN(concate_dnn_hidden_units, nn.ReLU(), 0.0, 0.0, False)
        self.host_concate_dnn = DNN(concate_dnn_hidden_units, nn.ReLU(), 0.0, 0.0, False)
        
        self.user_dnn = DNN(user_dnn_hidden_units, nn.ReLU(), 0.0, 0.0, False)

        #gate
        self.gate=Gate(weight_dnn_hidden_units)

    def forward(self,house_continuous_features,house_host_continuous_features,house_categorical_features1,house_categorical_feature2,house_multi_features,house_comment_features,
                 host_continuous_features, host_categorical_features1, host_categorical_features2,host_comment_features, host_pic_features,
                 user_house_continuous_features,user_house_host_continuous_features,user_house_categorical_features1,user_house_categorical_feature2,user_house_multi_features,user_house_comment_features,
                 user_host_continuous_features,user_host_categorical_features1,user_host_categorical_features2,user_host_comment_features,user_host_pic_features
                ,mask):
        #house
        # shape(样本数,2,200)
        house_vec=self.house_embedding(house_continuous_features,house_categorical_features1,house_categorical_feature2,house_multi_features)
        house_host_vec=self.house_host_embedding(house_host_continuous_features,house_comment_features)
        
        # shape(样本数, 15,200)
        user_house_emb=self.house_embedding(user_house_continuous_features,user_house_categorical_features1,user_house_categorical_feature2,user_house_multi_features)
        user_house_host_emb=self.house_host_embedding(user_house_host_continuous_features,user_house_comment_features)
        
        #分解 house_host_vec
        house_mmoe_outputs=self.mmoe_model(house_host_vec)
        v1=house_mmoe_outputs[0]  ##shape(样本数, 6,200)
        v2 = house_mmoe_outputs[1]
        v1_emb=self.mmoe_mlp1(v1)#shape(样本数, 6,200)
        v2_emb=self.mmoe_mlp2(v2)
        
        #分解user_house_emb
        user_house_mmoe_outputs = self.mmoe_model(user_house_host_emb)
        user_v1 = user_house_mmoe_outputs[0]  #shape(样本数, 15,200)
        user_v2 = user_house_mmoe_outputs[1]  #shape(样本数, 15,200)
        user_v1_emb=self.mmoe_mlp1(user_v1)#shape(样本数,15,200)
        user_v2_emb=self.mmoe_mlp2(user_v2)
        
        #历史记录中的15个house和正负样本house一起经过鉴别器训练
        all_house_host_emb1 = torch.cat([v1_emb, user_v1_emb], dim=-2)  # [样本数, 17, 200]
        all_house_host_emb2 = torch.cat([v2_emb, user_v2_emb], dim=-2)  # [样本数, 17, 200]
        all_house=torch.cat([house_vec, user_house_emb], dim=-2)  # [样本数, 17, 200]
        
        
        #合并house
        house_concate=torch.cat([house_vec, v1_emb], dim=-1)# [样本数, 6, 400]
        user_house_concate=torch.cat([user_house_emb, user_v1_emb], dim=-1)# [样本数, 15, 400]
        house=self.house_concate_dnn(house_concate)# [样本数, 6, 200]
        user_house=self.house_concate_dnn(user_house_concate)# [样本数, 15, 200]
        #house用户建模
        user_house_vec=self.user_house_encoder(user_house, mask)  #shape(样本数, 200)


        #host
        #host_continuous_features(样本数，6,12)
        #host_comment_features(样本数，6,10)
        #host_pic_features(样本数，6,159)
        # host_categories_embeddings1，shape(样本数，2,10(2*5))
        host_vec=self.host_embedding(host_continuous_features, host_categorical_features1, host_categorical_features2,host_comment_features, host_pic_features)
        user_host_emb=self.host_embedding(user_host_continuous_features,user_host_categorical_features1,user_host_categorical_features2,user_host_comment_features,user_host_pic_features)
        
        all_host=torch.cat([host_vec, user_host_emb], dim=-2)  # [样本数, 15+6, 200]
        
        #合并host
        host_concate=torch.cat([host_vec, v2_emb], dim=-1)# [样本数, 6, 400]
        user_host_concate=torch.cat([user_host_emb, user_v2_emb], dim=-1)# [样本数, 15, 400]
        host=self.host_concate_dnn(host_concate)
        user_host=self.host_concate_dnn(user_host_concate)       
        
        #host用户建模
        user_host_vec = self.user_host_encoder(user_host, mask)  # shape(样本数, 200)

        #历史记录中的15个host和正负样本host一起做鉴别器label
#         all_host_emb = torch.cat([host_vec, user_host_emb], dim=-2)  # [样本数, 15+6, 200]




        #自适应权重
        user_vec=torch.cat([user_house_vec,user_host_vec],dim=-1) #(样本数,400)
        user_input=self.user_dnn(user_vec)  #[样本数,1， 200]
        item_vec = torch.cat([house, host], dim=-1) #(样本数,6,400)
        gate_weight = self.gate(user_input,item_vec) #(样本数,6, 2)

        #匹配分数
        score1 = torch.sigmoid(torch.matmul(house, user_house_vec.unsqueeze(-1)))
        score2 = torch.sigmoid(torch.matmul(host,  user_host_vec.unsqueeze(-1)))

        # 转换为 (样本数, 2) 的维度
        score1 = score1.squeeze(-1)
        score2 = score2.squeeze(-1)

        #加权平均
        scores = torch.stack((score1, score2), dim=2)  #[样本数, 6, 3]
        weighted_scores = torch.mul(gate_weight, scores)   #[样本数, 6, 2]
        # 在最后一维上求和，得到每一对样本的加权分数
        sum_scores = torch.sum(weighted_scores, dim=-1)  #[样本数, 6]


        return sum_scores,all_house_host_emb1,all_house_host_emb2,all_house,all_host,scores,gate_weight,house,host
```

```python
from torch.utils.data import Dataset, DataLoader, TensorDataset



class BPRLoss(nn.Module):
    def __init__(self):
        super(BPRLoss, self).__init__()

    def forward(self, positive_scores, negative_scores):
        
        # 计算正样本和负样本之间的分数差异
        loss = -torch.log(torch.sigmoid(positive_scores - negative_scores))  #positive_scores自动进行广播
        return torch.mean(loss)


```

```python
def test_model(model, test_loader):
    model.eval()  # 设置模型为评估模式
    with torch.no_grad():  # 在评估模式下不计算梯度
        total_test_loss = 0.0
        total_rec_loss = 0.0
        house_total_rec_loss = 0.0
        host_total_rec_loss = 0.0
        total_regu_loss_test=0.0
        total_test_auc=0.0
        house_total_test_auc=0.0
        host_total_test_auc=0.0
        total_distance_correlation_v1_v2=0.0
        total_distance_correlation_v1_host=0.0
        total_distance_correlation_v1_house=0.0
        total_distance_correlation_v2_host=0.0
        total_distance_correlation_v2_house=0.0
        total_distance_correlation_house_host=0.0
        total_distance_correlation_house_host_raw=0.0
        results = []  # 用于保存结果的列表
        for batch_test in test_loader:  # 假设你有一个名为 val_loader 的验证集数据加载器
            batch_test = [data.to(device) for data in batch_test]
            house,host,reviewer,house_continuous_features_test,house_host_continuous_features_test,house_categorical_features1_test, house_categorical_feature2_test, house_multi_features_test, house_comment_features_test, \
            host_continuous_features_test, host_categorical_features1_test, host_categorical_features2_test, host_comment_features_test, host_pic_features_test, \
            user_house_continuous_features_test,user_house_host_continuous_features_test, user_house_categorical_features1_test, user_house_categorical_feature2_test, user_house_multi_features_test, user_house_comment_features_test, \
            user_host_continuous_features_test, user_host_categorical_features1_test, user_host_categorical_features2_test, user_host_comment_features_test, user_host_pic_features_test, \
            label_test, mask_test = batch_test
                    
            sum_scores_test,all_house_host_emb1_test,all_house_host_emb2_test,all_house_test,all_host_test,scores_test,gate_weight_test,house_test,host_test= model(house_continuous_features_test,house_host_continuous_features_test,house_categorical_features1_test,
                                                                                                                            house_categorical_feature2_test, house_multi_features_test, 
                                                                                                                            house_comment_features_test, host_continuous_features_test, 
                                                                                                                            host_categorical_features1_test, host_categorical_features2_test, 
                                                                                                                            host_comment_features_test, host_pic_features_test,                                                                                           
                                                                                                                            user_house_continuous_features_test,user_house_host_continuous_features_test,user_house_categorical_features1_test, 
                                                                                                                            user_house_categorical_feature2_test, user_house_multi_features_test,
                                                                                                                            user_house_comment_features_test, user_host_continuous_features_test, 
                                                                                                                            user_host_categorical_features1_test, user_host_categorical_features2_test, 
                                                                                                                            user_host_comment_features_test, user_host_pic_features_test,mask_test)
            # 计算验证集上的损失
            # 对抗损失
#             att_loss1_test = att_criterion(att_output1_test, all_host_emb_test)  # 损失函数越大越好
#             att_loss2_test = att_criterion(att_output2_test, all_host_emb_test)  # 损失函数越小越好
#             att_loss_test = -torch.log(torch.sigmoid(att_loss1_test - att_loss2_test))

            
            #计算三个向量间的余弦相似度，（batch，17）
            similarity_v1_v2_test=F.cosine_similarity(all_house_host_emb1_test,all_house_host_emb2_test, dim=2)
            similarity_v1_host_test=F.cosine_similarity(all_house_host_emb1_test,all_host_test, dim=2)
            similarity_v1_house_test=F.cosine_similarity(all_house_host_emb1_test,all_house_test, dim=2)
            similarity_v2_host_test=F.cosine_similarity(all_house_host_emb2_test,all_host_test, dim=2)
            similarity_v2_house_test=F.cosine_similarity(all_house_host_emb2_test,all_house_test, dim=2)
            
            #batch整体相关性
            distance_correlation_v1_v2_test=create_distance_correlation(all_house_host_emb1_test,all_house_host_emb2_test)  #一个值
            distance_correlation_v1_host_test=create_distance_correlation(all_house_host_emb1_test,all_host_test)
            distance_correlation_v1_house_test=create_distance_correlation(all_house_host_emb1_test,all_house_test)
            distance_correlation_v2_host_test=create_distance_correlation(all_house_host_emb2_test,all_host_test)
            distance_correlation_v2_house_test=create_distance_correlation(all_house_host_emb2_test,all_house_test)

            # 约束1
            regu_loss1_test= torch.mean(distance_correlation_v1_host_test)
            # 约束2
            regu_loss2_test=torch.mean(1-distance_correlation_v2_host_test)
            # 约束3
            regu_loss3_test=torch.mean(distance_correlation_v2_house_test)
            #总约束
            regu_loss_test=λ1*regu_loss1_test+λ2*regu_loss2_test+λ3*regu_loss3_test
            
            #推荐损失
            positive_scores_test = sum_scores_test[:, 0]
            positive_scores_test = positive_scores_test.unsqueeze(-1)
            negative_scores_test = sum_scores_test[:, 1:]
            rec_loss_test = rec_criterion(positive_scores_test, negative_scores_test)
            #house推荐损失
            house_scores_test=scores_test[:,:,0]  #(batch,2)
            house_positive_scores_test = house_scores_test[:, 0]
            house_negative_scores_test = house_scores_test[:, 1]
            house_rec_loss_test = rec_criterion(house_positive_scores_test, house_negative_scores_test)            
            #host推荐损失
            host_scores_test=scores_test[:,:,1] #(batch,2)
            host_positive_scores_test = host_scores_test[:, 0]
            host_negative_scores_test = host_scores_test[:, 1]
            host_rec_loss_test = rec_criterion(host_positive_scores_test, host_negative_scores_test)            
            #总损失
            test_loss = rec_loss_test + λ*regu_loss_test
            total_test_loss += test_loss.item()
            total_rec_loss += rec_loss_test.item()
            house_total_rec_loss += house_rec_loss_test.item()
            host_total_rec_loss += host_rec_loss_test.item()
            total_regu_loss_test+=regu_loss_test.item()
            
            # 计算验证集上的AUC
            predictions_test = (positive_scores_test > negative_scores_test).float()
            correct_test = predictions_test.sum().item()
            total_test_auc += correct_test / (len(predictions_test)*npratio)
            
            # house的AUC
            house_predictions_test = (house_positive_scores_test > house_negative_scores_test).float()
            house_correct_test = house_predictions_test.sum().item()
            house_total_test_auc += house_correct_test / len(house_predictions_test)
            
            # host的AUC
            host_predictions_test = (host_positive_scores_test > host_negative_scores_test).float()
            host_correct_test = host_predictions_test.sum().item()
            host_total_test_auc += host_correct_test / len(host_predictions_test)

            #计算三个向量间的余弦相似度，（batch，2）
            similarity_house_host=F.cosine_similarity(house_test,host_test, dim=2)
            similarity_house_host_raw=F.cosine_similarity(all_house_test,all_host_test, dim=2)

            #batch整体相关性
            distance_correlation_house_host=create_distance_correlation(house_test,host_test)  #一个值
            distance_correlation_house_host_raw=create_distance_correlation(all_house_test,all_host_test)

            total_distance_correlation_v1_v2+=distance_correlation_v1_v2_test.item()
            total_distance_correlation_v1_host+=distance_correlation_v1_host_test.item()
            total_distance_correlation_v1_house+=distance_correlation_v1_house_test.item()
            total_distance_correlation_v2_host+=distance_correlation_v2_host_test.item()
            total_distance_correlation_v2_house+=distance_correlation_v2_house_test.item()
            total_distance_correlation_house_host+=distance_correlation_house_host.item()
            total_distance_correlation_house_host_raw+=distance_correlation_house_host_raw.item()
       
            weighted_scores_test=F.softmax(torch.mul(gate_weight_test, scores_test), dim=-1)   #[样本数, 2, 2]

            for i in range(house.shape[0]):
                #正样本
                results.append({
                    'house': house[i][0],
                    'host': host[i][0],
                    'reviewer': reviewer[i],
                    'sum_score': sum_scores_test[i][0],  # 正样本分数
                    'score1': scores_test[i][0][0],  # 三部分分数
                    'score2': scores_test[i][0][1],

                    'weight1': gate_weight_test[i][0][0],  # 正样本权重
                    'weight2': gate_weight_test[i][0][1],
                    'weighted_scores_test1': weighted_scores_test[i][0][0],
                    'weighted_scores_test2': weighted_scores_test[i][0][1],

                    'similarity_v1_v2':similarity_v1_v2_test[i][0],
                    'similarity_v1_host':similarity_v1_host_test[i][0],
                    'similarity_v1_house':similarity_v1_house_test[i][0],
                    'similarity_v2_host':similarity_v2_host_test[i][0],
                    'similarity_v2_house':similarity_v2_house_test[i][0],
                    'similarity_house_host_raw':similarity_house_host_raw[i][0],
                    'similarity_house_host':similarity_house_host[i][0],
                    'label': label_test[i][0]
                    
                })
                for j in range(npratio):
                    #负样本
                    results.append({
                        'house': house[i][1+j],
                        'host': host[i][1+j],
                        'reviewer': reviewer[i],             
                        'sum_score': sum_scores_test[i][1+j],  # 负样本分数
                        'score1': scores_test[i][1+j][0],  # 三部分分数
                        'score2': scores_test[i][1+j][1],

                        'weight1': gate_weight_test[i][1+j][0],  # 正样本权重
                        'weight2': gate_weight_test[i][1+j][1],
                        'weighted_scores_test1': weighted_scores_test[i][1+j][0],
                        'weighted_scores_test2': weighted_scores_test[i][1+j][1],

                        'similarity_v1_v2':similarity_v1_v2_test[i][1+j],
                        'similarity_v1_host':similarity_v1_host_test[i][1+j],
                        'similarity_v1_house':similarity_v1_house_test[i][1+j],
                        'similarity_v2_host':similarity_v2_host_test[i][1+j],
                        'similarity_v2_house':similarity_v2_host_test[i][1+j],
                        'similarity_house_host_raw':similarity_house_host_raw[i][1+j],
                        'similarity_house_host':similarity_house_host[i][1+j],
                        'label': label_test[i][1+j]
                    })
                    
        average_test_loss = total_test_loss / len(test_loader)
        average_auc_test=total_test_auc/ len(test_loader)
        house_average_auc_test=house_total_test_auc/ len(test_loader)
        host_average_auc_test=host_total_test_auc/ len(test_loader)
        average_rec_loss = total_rec_loss / len(test_loader)
        house_average_rec_loss = house_total_rec_loss / len(test_loader)
        host_average_rec_loss = host_total_rec_loss / len(test_loader)
        average_regu_loss_test=total_regu_loss_test/ len(test_loader)
        average_distance_correlation_v1_v2=total_distance_correlation_v1_v2/ len(test_loader)
        average_distance_correlation_v1_host=total_distance_correlation_v1_host/ len(test_loader)
        average_distance_correlation_v1_house=total_distance_correlation_v1_house/ len(test_loader)
        average_distance_correlation_v2_host=total_distance_correlation_v2_host/ len(test_loader)
        average_distance_correlation_v2_house=total_distance_correlation_v2_house/ len(test_loader)
        average_distance_correlation_house_host=total_distance_correlation_house_host/ len(test_loader)
        average_distance_correlation_house_host_raw=total_distance_correlation_house_host_raw/ len(test_loader)
        print(f"Test Loss: {average_test_loss},rec_loss:{average_rec_loss},AUC: {average_auc_test},att_loss:{average_regu_loss_test}")
        print(f"house_rec_loss:{house_average_rec_loss},house_AUC: {house_average_auc_test}")
        print(f"host_rec_loss:{host_average_rec_loss},host_AUC: {host_average_auc_test}")
        
        print(f"average_distance_correlation_v1_v2: {average_distance_correlation_v1_v2},average_distance_correlation_v1_host:{average_distance_correlation_v1_host},average_distance_correlation_v1_house:{average_distance_correlation_v1_house},average_distance_correlation_v2_host:{average_distance_correlation_v2_host},average_distance_correlation_v2_house:{average_distance_correlation_v2_house},average_distance_correlation_house_host:{average_distance_correlation_house_host},average_distance_correlation_house_host_raw:{average_distance_correlation_house_host_raw}")
        
        
        # 将结果列表转换为DataFrame
        df_results = pd.DataFrame(results)
        df_results['house'] = df_results['house'].apply(lambda x: x.item())
        df_results['host'] = df_results['host'].apply(lambda x: x.item())
        df_results['reviewer'] = df_results['reviewer'].apply(lambda x: x.item())
        df_results['sum_score'] = df_results['sum_score'].apply(lambda x: x.item())
        df_results['score1'] = df_results['score1'].apply(lambda x: x.item())
        df_results['score2'] = df_results['score2'].apply(lambda x: x.item())

        df_results['weight1'] = df_results['weight1'].apply(lambda x: x.item())
        df_results['weight2'] = df_results['weight2'].apply(lambda x: x.item())
        
        df_results['weighted_scores_test1'] = df_results['weighted_scores_test1'].apply(lambda x: x.item())
        df_results['weighted_scores_test2'] = df_results['weighted_scores_test2'].apply(lambda x: x.item())

        df_results['similarity_v1_v2'] = df_results['similarity_v1_v2'].apply(lambda x: x.item())
        df_results['similarity_v1_host'] = df_results['similarity_v1_host'].apply(lambda x: x.item())
        df_results['similarity_v1_house'] = df_results['similarity_v1_house'].apply(lambda x: x.item())
        df_results['similarity_v2_host'] = df_results['similarity_v2_host'].apply(lambda x: x.item())
        df_results['similarity_v2_house'] = df_results['similarity_v2_house'].apply(lambda x: x.item())
        df_results['similarity_house_host_raw'] = df_results['similarity_house_host_raw'].apply(lambda x: x.item())
        df_results['similarity_house_host'] = df_results['similarity_house_host'].apply(lambda x: x.item())
        df_results['label'] = df_results['label'].apply(lambda x: x.item())
        df_results['item_count'] = df_results.groupby('reviewer')['reviewer'].transform('count')
        
  
        
        
        return df_results,average_test_loss,average_auc_test,house_average_auc_test,host_average_auc_test,average_rec_loss,house_average_rec_loss,host_average_rec_loss,average_regu_loss_test,average_distance_correlation_v1_v2,average_distance_correlation_v1_host,average_distance_correlation_v1_house,average_distance_correlation_v2_host,average_distance_correlation_v2_house,average_distance_correlation_house_host,average_distance_correlation_house_host_raw
```

```python
#top-n
def top_evaluation(df_results,n):
    # 初始化结果列表
    ndcg_list = []
    hr_list=[]
    
    for user_id, user_group in df_results.groupby('reviewer'):
        user_group = user_group.sort_values(by='sum_score', ascending=False)  # 按分数降序排序
        top_items = user_group.head(n)['label'].tolist()  # 选取前n个的标签
        true_labels = user_group['label'].tolist()  # 用户所有候选物品的真实标签
                
        # 计算NDCG
        idcg = sorted(true_labels, reverse=True)[:n]  # 理想情况下的DCG
        dcg = sum(rel / np.log(i + 2) for i, rel in enumerate(top_items))  # 计算DCG
        ndcg = dcg / sum(rel / np.log(i + 2) for i, rel in enumerate(idcg))  # 计算NDCG
        ndcg_list.append(ndcg)
        # 计算HR@N
        hit_rate = 1 if any(label == 1 for label in top_items) else 0  # 若Top N中有真实正样本，HR@N为1，否则为0
        hr_list.append(hit_rate)

        
    # 计算平均值
    average_ndcg = np.mean(ndcg_list)
    average_hr = np.mean(hr_list)
    
    return average_ndcg, average_hr
```
